```python
# MELT & PIVOT 
df= pd.DataFrame({
    "country" : ["usa" ,"usa" ,"india" ,"india"],
    "year" : [2020,2021,2020,2021],
    "sales" :[100,120,90,110],
    "profit" : [20,25,18,22]
})


melted_df=df.melt(
   id_vars = ["country" ,"year"],
   value_vars=["sales","profit"],
    var_name="matrics",
    value_name="value"
)

melted_df.pivot(
    index=["country","year"],
    columns="matrics",
    values="value"
 )

```

```python
# Merging & Joining data 

df_customers= pd.DataFrame({
    "customer_id" : [1,2,3,4],
    "name" : ["Adam" , "Bob" ,"Charlie" ,"Dave"]
})

df_orders = pd.DataFrame({
    "order_id" : [101,102,103,104],
    "customer_id": [2,1,4,5],
    "amount": [250,120,300,180]
})


pd.merge(df_customers,df_orders, on= "customer_id")
pd.merge(df_customers,df_orders, on= "customer_id",how="left")
pd.merge(df_customers,df_orders, on= "customer_id",how="right")
pd.merge(df_customers,df_orders, on= "customer_id",how="outer")
```
