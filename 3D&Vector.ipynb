```python
import numpy as np 
import time
```

```python
#brodcasting 
arr1= np.array([1,2,3,4,5])
arr2= np.array([[1,2,3,4,5],[5,6,7,8,9]])
print(arr1.shape,arr2.shape)
print(arr1+arr2)
```

```python
#normalize
arr = np.array([[1,2],[3,4]])

mean = np.mean(arr)
std_dev= np.std(arr)

normalized_arr =(arr-mean)/std_dev
print(normalized_arr)
print(mean)
print(std_dev)
```

```python
arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.argmin(arr))
print(np.argmax(arr))
print(np.mean(arr))
print(np.std(arr))
```

```python
arr = np.array([1,2,3,4,5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
print(np.log(arr))
print(np.exp(arr))
print(np.log2(arr))
```

```python
#rounding 
print(np.round(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))

arr =np.array([1,-2,8,-3,4,5,2,5])

print(np.unique(arr))
print(np.round(3.14))
print(np.sort(arr))
print(np.abs(arr))
```
